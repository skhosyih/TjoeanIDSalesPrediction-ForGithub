```python
# Load datasets and convert it into excel based on month of sales's sheets
JAN21 = pd.read_excel(r"..\..\data\dataCleaning\JAN21_dataCleaning.xlsx")
FEB21 = pd.read_excel(r"..\..\data\dataCleaning\FEB21_dataCleaning.xlsx")
MAR21 = pd.read_excel(r"..\..\data\dataCleaning\MAR21_dataCleaning.xlsx")
APR21 = pd.read_excel(r"..\..\data\dataCleaning\APR21_dataCleaning.xlsx")
MAY21 = pd.read_excel(r"..\..\data\dataCleaning\MAY21_dataCleaning.xlsx")
JUN21 = pd.read_excel(r"..\..\data\dataCleaning\JUN21_dataCleaning.xlsx")
JUL21 = pd.read_excel(r"..\..\data\dataCleaning\JUL21_dataCleaning.xlsx")
AUG21 = pd.read_excel(r"..\..\data\dataCleaning\AUG21_dataCleaning.xlsx")
SEP21 = pd.read_excel(r"..\..\data\dataCleaning\SEP21_dataCleaning.xlsx")
OCT21 = pd.read_excel(r"..\..\data\dataCleaning\OCT21_dataCleaning.xlsx")
NOV21 = pd.read_excel(r"..\..\data\dataCleaning\NOV21_dataCleaning.xlsx")
DEC21 = pd.read_excel(r"..\..\data\dataCleaning\DEC21_dataCleaning.xlsx")

JAN22 = pd.read_excel(r"..\..\data\dataCleaning\JAN22_dataCleaning.xlsx")
FEB22 = pd.read_excel(r"..\..\data\dataCleaning\FEB22_dataCleaning.xlsx")
MAR22 = pd.read_excel(r"..\..\data\dataCleaning\MAR22_dataCleaning.xlsx")
APR22 = pd.read_excel(r"..\..\data\dataCleaning\APR22_dataCleaning.xlsx")
MAY22 = pd.read_excel(r"..\..\data\dataCleaning\MAY22_dataCleaning.xlsx")
JUN22 = pd.read_excel(r"..\..\data\dataCleaning\JUN22_dataCleaning.xlsx")
JUL22 = pd.read_excel(r"..\..\data\dataCleaning\JUL22_dataCleaning.xlsx")
AUG22 = pd.read_excel(r"..\..\data\dataCleaning\AUG22_dataCleaning.xlsx")
SEP22 = pd.read_excel(r"..\..\data\dataCleaning\SEP22_dataCleaning.xlsx")
OCT22 = pd.read_excel(r"..\..\data\dataCleaning\OCT22_dataCleaning.xlsx")
NOV22 = pd.read_excel(r"..\..\data\dataCleaning\NOV22_dataCleaning.xlsx")
DEC22_DEC23 = pd.read_excel(r"..\..\data\dataCleaning\DEC22-DEC23_dataCleaning.xlsx")
```

### Concatenate all the datasets

```python
df = pd.concat([JAN21, FEB21, MAR21, APR21, MAY21, JUN21, JUL21, AUG21, SEP21, OCT21, NOV21, DEC21,
                JAN22, FEB22, MAR22, APR22, MAY22, JUN22, JUL22, AUG22, SEP22, OCT22, NOV22, DEC22_DEC23], ignore_index=True, axis=0)
```

```python
df.info()
df
```

```python
# Convert 'DATE' into datetime
df['DATE'] = pd.to_datetime(df['DATE'])
```

```python
# Sort values based on 'DATE' and reset the index
df = df.sort_values(by=['DATE']).reset_index(drop=True)

df.info();
df.head()
```

```python
def create_features(df):
    """
    Create time series features based on time series index.
    """
    df = df.copy()
    # df['DayOfWeek'] = df.dayofweek
    # df['Quarter'] = df.index.quarter
    # df['DayOfMonth'] = df.index.day
    # df['Day'] = df['DATE'].dt.day
    # df['Month'] = df['DATE'].dt.month
    df['Month'] = df.index + 1
    # df['Year'] = df['DATE'].dt.year
    # df['DayOfYear'] = df.index.dayofyear  
    # df['WeekOfYear'] = df.index.isocalendar().week  
    
    # Change datatypes
    # df['Day'] = df['Day'].astype('int64')
    df['Month'] = df['Month'].astype('int64')
    # df['Year'] = df['Year'].astype('int64')

    return df

df = create_features(df)
```
